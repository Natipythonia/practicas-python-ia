````python
var_1 = [0, 1]
var_2 = [0,1,2,3,4,5]
if len(var_1) > len(var_2): 
    print("var_1 tiene longitud mayor que var_2")
elif len(var_1) == len(var_2):
    print("var_1 tiene longitud igual que var_2")
elif var_1[0] == var_2[0]:
    print("Ambas listas comienzan por el mismo valor")
else:
    print("No se cumple ninguna de las condiciones")
````

## 12) Problema final
Contexto:

Estás trabajando brindando apoyo en el desarrollo de un subsistema informático de una empresa de venta de seguros. Tu responsable necesita que hagas un programa sencillo que transforme una lista de coberturas en una cadena de texto para mostrársela al usuario.

Requisitos:

- Omitir campos con valor None: Si algún campo en los datos de la cobertura tiene el valor None, debes omitir esa parte en el texto resultante.
- Valor especial para LIMMAXRESP: Si el valor de "LIMMAXRESP" es igual a "1", debes reemplazarlo por "el valor establecido por la ley".


Datos del ejemplo proporcionado por tu responsable:
```
listado_coberturas = [
    {"cobertura_1": {
        "COBERTURA": "Robo",
        "LIMMAXRESP": "5000",
        "PRIMAS": "200"
    }},
    {"cobertura_2": {
        "COBERTURA": "Incendio",
        "LIMMAXRESP": "1",
        "PRIMAS": None
    }}
]
```

Formato esperado del texto resultante a partir de los datos anteriores:
```
Las coberturas disponibles para el pack elegido son las siguientes:
La cobertura nº1 cubre "Robo" con un límite máximo de 5000€ con una prima de 200€.
La cobertura nº2 cubre "Incendio" con un límite máximo del valor establecido por la ley.
```

````python
listado_coberturas = [
    {"cobertura_1": {
        "COBERTURA": "Robo",
        "LIMMAXRESP": "5000",
        "PRIMAS": "200"
    }},
    {"cobertura_2": {
        "COBERTURA": "Incendio",
        "LIMMAXRESP": "1",
        "PRIMAS": None
    }}
]
````

````python
def formatea_cobertura(indice, cobertura_dict):
    """
    Devuelve una frase con el formato:
    La cobertura nº{indice} cubre "{COBERTURA}" con un límite máximo {de X€ / del valor establecido por la ley} {y con una prima de Y€}.
    Omite las partes cuyo valor sea None.
    """
    # El dict viene envuelto como {"cobertura_1": {...}} → nos quedamos con el valor interno
    datos = next(iter(cobertura_dict.values()))

    nombre = datos.get("COBERTURA")
    lim = datos.get("LIMMAXRESP")
    primas = datos.get("PRIMAS")

    # Vamos con la frase base
    partes = [f'La cobertura nº{indice} cubre "{nombre}"']

    # Límite máximo (omitir si es None)
    if lim is not None: 
        if str(lim) == "1":
            partes.append("con un límite máximo de valor establecido por la ley")
        else: 
            partes.append(f"con un límite máximo de {lim}€")

    # Primas (omitir si es None)
    if primas is not None:
        partes.append(f"con una prima de {primas}€")

    # Unimos todas las partes con espacios y cerramos con punto
    return " ".join(partes) + "."
````
