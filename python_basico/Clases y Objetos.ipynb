## 1) Crear una clase básica con atributos y métodos básicos
Escribe un programa en Python para crear una clase Vehicle con atributos de instancia __max_speed__ , __mileage__ , __lights__ y __owner__. 

Y 4 métodos:
<ul>
    <li>get_max_speed() -> Tiene que devolver la velocidad máxima</li>

</ul>

Instancia al menos una vez el objeto, pon tu nombre y apellidos como propietario del vehículo y llama a cada uno de los métodos al menos una vez.

```python
class Vehicle:
    """
    Clase que representa un vehículo con velocidad máxima, kilometraje,
    luces y propietario.
    """

    def __init__(self, max_speed, mileage, lights=False, owner="Desconocido"):
        # Atributos de instancia
        self.max_speed = max_speed      # Velocidad máxima
        self.mileage = mileage          # Kilometraje
        self.lights = lights            # Estado de las luces (True/False)
        self.owner = owner              # Propietario del vehículo

    def get_max_speed(self):
        """Devuelve la velocidad máxima del vehículo."""
        return self.max_speed

    def set_lights_on(self):
        """Enciende las luces del vehículo."""
        self.lights = True

    def increase_mileage(self, amount_to_increase):
        """Incrementa el kilometraje según el valor recibido."""
        if amount_to_increase < 0:
            print("El incremento no puede ser negativo.")
        else:
            self.mileage += amount_to_increase

    def display_owner(self):
        """Muestra por pantalla el propietario del vehículo."""
        print(f"El propietario del vehículo es: {self.owner}")


# ----------- PRUEBA DEL CÓDIGO -----------

# Creamos un vehículo con tus datos
my_car = Vehicle(max_speed=180, mileage=25000, owner="Nati Hernández Pérez")

# Llamamos a cada método
print("Velocidad máxima:", my_car.get_max_speed())  # -> 180

my_car.set_lights_on()
print("Luces encendidas:", my_car.lights)           # -> True

my_car.increase_mileage(500)
print("Nuevo kilometraje:", my_car.mileage)         # -> 25500

my_car.display_owner()                              # -> Muestra tu nombre
```

```python
class Rectangle: 
    """
    Clase que representa un rectángulo con longitud y anchura.
    """
    def __init__(self, length, width):
        """Inicializa los atributos del rectángulo."""
        self.length = length
        self.width = width
    def perimeter(self):
        """Calcula y devuelve el perímetro del rectángulo."""
        return 2 * (self.length + self.width)
    def area(self):
        """Calcula y devuelve el área del rectángulo."""
        return self.length * self.width
    def display(self): 
        """Devuelve una cadena con toda la información del rectángulo."""
        return (f"Longitud: {self.length}, "
                f"Anchura: {self.width}, "
                f"Perímetro: {self.perimeter()}, "
                f"Área: {self.area()}")
    def __str__(self): 
        """Sobreescribe la representación por defecto de la clase."""
        return self.display()

# Prueba el código
# Creamos una instancia del rectángulo
rect1 = Rectangle(length=10, width=5)

# Llamamos a cada método
print("Perímetro:", rect1.perimeter()) # -> 30
print("Área:", rect1.area())           # -> 50
print(rect1.display())                 # -> Muestra todos los datos
print(rect1)                           # -> Usa __str__, muestra igual que display()
```

```python
class BankAccount:
    """
    Clase que representa una cuenta bancaria con operaciones básicas.
    """

    def __init__(self, account_number, name, balance=0.0):
        """Constructor con número de cuenta, titular y saldo inicial."""
        self.account_number = account_number
        self.name = name
        self.balance = float(balance)

    # Métodos principales

    def deposit(self, amount): 
        """Realiza un depósito en la cuenta."""
        if amount <= 0:
            print("El depósito debe ser mayor que cero.")
        else:
            self.balance += amount
            print(f"Se han depositado {amount}€. Nuevo saldo: {self.balance}€")

    def withdrawal(self, amount):
        """Realiza una retirada de dinero de la cuenta."""
        if amount <= 0:
            print("La cantidad a retirardebe ser mayor que cero.")
        elif amount > self.balance:
            print("Fondos insuficientes.")
        else:
            self.balance -= amount
            print(f"Se han retirado {amount}€. Saldo restante: {self.balance}€")

    def transfer(self, acc_to_transfer_to_obj, amount):
        """Transfiere dinero a otra cuenta."""
        if amount <= 0:
            print("La cantidad a transferir debe ser mayor que cero.")
        elif amount > self.balance: 
            print("No hay suficiente saldo para transferir.")
        else:
            self.balance -= amount
            acc_to_transfer_to_obj.balance += amount
            print(f"Transferencia de {amount}€ realizada de {self.name} a {acc_to_transfer_to_obj.name}")

    def apply_bank_fees(self, percentage):
        """Aplica comisiones bancarias sobre el saldo."""
        if percentage < 0:
            print("El porcentaje no puede ser negativo.")
        else:
            fee = self.balance * (percentage / 100)
            self.balance -= fee
            print(f"Se aplicó una comisión del {percentage}%. Comisión: {fee}€. Nuevo saldo: {self.balance}€")

    def display (self):
        """Muestra los detalles de la cuenta."""
        print(f"Número de cuenta: {self.account_number}")
        print(f"Titular: {self.name}")
        print(f"Saldo: {self.balance:.2f}€")

    # SOBRECARGA DE OPERTADORES

    def __eq__(self, other):
        return self.balance == other.balance

    def __lt__(self, other):
        return self.balance < other.balance
        
    def __gt__(self, other): 
        return self.balance > other.balance

    def __ge__(self, other):
        return self.balance >= other.balance
```

```python
class Vehicle: 
    """
    Clase base para vehículos.
    Tarifas: capacidad * 110
    Multas: lista de dicts {'fecha': str, 'cuantía': float}
    """
    def __init__(self, name: str, mileage: float, capacity: int):
        self.name = name
        self.mileage = mileage
        self.capacity = capacity
        self.multas = []  # <- nuevo atributo
    def fare(self) -> float: 
        return self.capacity * 110

    # --------- Gestión de multas ---------
    
    def anyadir_multa(self, fecha: str, cuantia: float):
        """Añade una multa con fecha (str) y cuantía (float > 0)."""
        try: 
            cuantia = float(cuantia)
        except (TypeError, ValueError):
            raise ValueError("La cuantía debe ser numérica.")

        if cuantia <= 0:
            raise ValueError("La cuantía debe ser mayor que 0.")

        self.multas.append({"fecha": fecha, "cuantia": cuantia})

    def eliminar_multa(self, indice: int):
        """Elimina la multa por indice (0, 1, 2...)."""
        if not self.multas:
            raise IndexError("No hay multas que eliminar.")

        if indice < 0 or indice >= len(self.multas):
            raise IndexError("Indice de multa fuera de rango.")
        self.multas.pop(indice)

    # (opcional) para ver el estado
    def listar_multas(self):
        """Devuelve una copia de la lista de multas."""
        return list(self.multas)
```

Si la entrada es válida, realiza el cálculo e imprime el resultado. A continuación, se pide al usuario que proporcione una nueva entrada, y así sucesivamente, hasta que el usuario introduzca la salida.


Una interacción podría ser así:

```python
# Excepción personalizada para errores de fórmula
class FormulaError(Exception): 
    """Excepción lanzada cuando la fórmula del usuario no es válida."""
    pass
```

```python
def calcular(formula: str) -> float:
    """
    Recibe una cadena con formato 'num1 operador nume2'.
    Valida y devuelve el resultado.
    Lanza FormulaError si el formato no es válido.
    """
    partes = formula.split()

    # Debe haber exactamente 3 elementos
    if len(partes) != 3: 
        raise FormulaError("La fórmula debe tener el formato: número operador numero")
    num1_str, operador, num2_str = partes

    # Intentar convertir los números 
    try:
        num1 = float(num1_str)
        num2 = float(num2_str)
    except ValueError:
        raise FormulaError("Los valores numéricos no son válidos")

    # Comprobar operador 
    if operador not in ['+', '-']:
        raise FormulaError("El operador debe ser '+' o '-'.")

    # Calcular resultado
    if operador == '+': 
        return num1 + num2
    else:
        return num1 - num2
```

Continúa con el ejercicio anterior y crea ahora tres clases hijas de FormulaError:

<ul>
<li><b>FormulaErrorNumber</b>: La levantarás cuando el usuario introduzca un valor erróneo para los números</li>
<li><b>FormulaErrorOp</b>: Se lanzará cuando el usuario introduzca un valor erróneo para el operador</li>
<li><b>FormulaErrorArgs</b>: Se lanzará cuando el usuario introduzca un número incorrecto de argumentos a la calculadora</li>
</ul>

Levántalos en los casos descritos y manéjalos con try, excepto para que puedas mostrar mensajes personalizados al usuario dependiendo de lo que esté mal en la entrada.

Puntos extra: Si también implementas la operación de división intenta manejar el caso ZeroDivisionError también


```python
# Excepciones personalizadas para la calculadora

class FormulaError(Exception):
    """Excepción base para los errores de la fórmula."""
    pass
    
class FormulaErrorNumber(FormulaError):
    """Error en los números introducidos."""
    pass

class FormulaErrorOp(FormulaError):
    """Error en el operador introducido."""
    pass

class FormulaErrorArgs(FormulaError):
    """Error en el número de argumentos de la fórmula."""
    pass
```

```python
def calcular(formula: str) -> float:
    """
    Recibe una cadena con formato 'num1 operador num2'.
    Valida y devuelve el resultado.
    Lanza excepciones específicas según el tipo de error.
    """
    partes = formula.split()

    # 1 Número incorrecto de argumentos
    if len(partes) !=3:
        raise FormulaErrorArgs("La fórmula debe tener el formato: número operador número")

    num1_str, operador, num2_str = partes

    # 2 Intentar convertir los números
    try:
        num1 = float(num1_str)
        num2 = float(num2_str)
    except ValueError:
        raiseFormulaErrorNumber(f"Valores numéricos no válidos: '{num1_str}' y/o '{num2_str}'.")

    # 3 Comprobar operador 
    if operador not in ['+', '-', '/']:
        raise FormulaErrorOp(f"Operador no válido: '{operador}'. Usa solo '+', '-' o '/'.")

    # 4 Calcular resultado (con manejo de división)
    if operador == '+':
        return num1 + num2
    elif operador == '-':
        return num1 - num2
    elif operador == '/': 
        return num1 / num2
```

```python
print("Calculadora mejorada (+, -, /). Escribe 'quit' para salir.")

while True:
    entrada = input(">>> ")
    if entrada.lower() == "quit":
        print("¡Hasta luego!")
        break

    try:
        resultado = calcular(entrada)
        print(resultado)

    except FormulaErrorNumber as e:
        print("X Error en los números:", e)
    except FormulaErrorOp as e:
        print("X Error en el operador:", e)
    except FormulaErrorArgs as e:
        print("X Error emn el número de argumentos:", e)
    except ZeroDivisionError:
        print("X Error: no se puede dividir entre cero.")
    except FormulaError as e:
        print("X Error genérico de fórmula:", e)
```

Imagina que formas parte de un equipo que se encarga de diseñar una pequeña aplicación de gestión de una tienda online. Esta tienda es muy polifacética así que vende __productos__ muy distintos. Los clientes pueden crear un __carrito__ de compra donde __agregan__ los productos que desean adquirir. Finalmente, el carrito se asocia a una __orden__ de compra. 
Vamos modelar estas clases utilizando orientación a objetos, aplicando herencia, pero sobre todo usando el principio de <a href='https://es.wikipedia.org/wiki/Composici%C3%B3n_de_objetos'>composición</a>.

__Extra__: Al procesar el pago tienes usar las cuentas bancarias. Para ello reutiliza la clase del ejercicio 3 considerando el pago como una transferencia de la cuenta del cliente a la cuenta de la empresa

```python
# ---------- Definición de clases ----------

class Producto:
    """
    Representa un producto de la tienda.
    """
    def __init__(self, nombre: str, precio: float):
        if not isinstance(nombre, str) or not nombre.strip():
            raise ValueError("El nombre del producto debe ser un string no vacío.")
        try:
            precio = float(precio)
        except (TypeError, ValueError):
            raise ValueError("El precio debe ser un número.")
        if precio < 0:
            raise ValueError("El precio no puede ser negativo.")

        self.nombre = nombre.strip()
        self.precio = precio

    def __str__(self) -> str:
        """Devuelve un string con la información del producto."""
        return f"{self.nombre} - {self.precio:.2f}€"
```

```python
class Carrito:
    """
    Representa un carrito compuesto de productos.
    """
    def __init__(self, productos=None):
        if productos is None:
            self.productos = []
        else:
            if not all(isinstance(p, Producto) for p in productos):
                raise TypeError("Todos los elementos deben ser instancias de Producto.")
            self.productos = list(productos)

    def agregar_producto(self, producto: Producto):
        """Agrega un producto al carrito."""
        if not isinstance(producto, Producto):
            raise TypeError("Solo se pueden agregar objetos de tipo Producto.")
        self.productos.append(producto)

    def eliminar_producto(self, producto: Producto):
        """Elimina el producto del carrito si existe."""
        if producto in self.productos:
            self.productos.remove(producto)

    def calcular_total(self) -> float:
        """Devuelve la suma de los precios de todos los productos."""
        return sum(p.precio for p in self.productos)

    def __str__(self) -> str:
        """Devuelve un string con el listado de productos y el total."""
        if not self.productos:
            return "Carrito vacío. Total: 0.00€"

        lineas = ["Carrito:"]
        for p in self.productos:
            lineas.append(f"  - {p}")
        lineas.append(f"Total: {self.calcular_total():.2f}€")
        return "\n".join(lineas)
```

```python
# ---------- PRUEBAS ----------

# Creamos algunos productos
p1 = Producto("Camiseta", 19.99)
p2 = Producto("Pantalón", 39.50)
p3 = Producto("Taza", 7.25)

print("Productos individuales:")
print(p1)
print(p2)
print(p3)
print()

# Creamos un carrito y probamos los métodos
carrito = Carrito()

# agregar_producto
carrito.agregar_producto(p1)
carrito.agregar_producto(p2)
carrito.agregar_producto(p3)

print("Carrito después de agregar productos:")
print(carrito)
print()

# calcular_total
print("Total calculado:", carrito.calcular_total())
print()

# eliminar_producto
carrito.eliminar_producto(p2)

print("Carrito después de eliminar un producto:")
print(carrito)
print("Nuevo total:", carrito.calcular_total())
```

```python
class Orden:
    """
    Representa una orden de compra.
    """
    def __init__(self, nombre_cliente: str, carrito: Carrito):
        if not isinstance(nombre_cliente, str) or not nombre_cliente.strip():
            raise ValueError("El nombre del cliente debe ser un string no vacío.")
        if not isinstance(carrito, Carrito):
            raise TypeError("El parámetro 'carrito' debe ser una instancia de Carrito.")

        self.nombre_cliente = nombre_cliente.strip()
        self.carrito = carrito
        self.estado = "En proceso"

    def procesar_pago(self):
        """Cambia el estado a 'Pagada' si el total del carrito es mayor que 0."""
        total = self.carrito.calcular_total()
        if total > 0:
            self.estado = "Pagada"
            print(f"Pago procesado correctamente. Total: {total:.2f}€")
        else:
            print("No se puede procesar el pago: el carrito está vacío.")

    def enviar_pedido(self):
        """Cambia el estado a 'Enviada' solo si el estado actual es 'Pagada'."""
        if self.estado == "Pagada":
            self.estado = "Enviada"
            print("El pedido ha sido enviado correctamente.")
        else:
            print("No se puede enviar: la orden aún no está pagada.")

    def __str__(self):
        """Devuelve información del cliente, estado y contenido del carrito."""
        return (
            f"Cliente: {self.nombre_cliente}\n"
            f"Estado: {self.estado}\n"
            f"{self.carrito}"
        )
```

```python
class Orden:
    """
    Representa una orden de compra.
    Ahora incluye las cuentas bancarias del cliente y de la empresa.
    """
    def __init__(self, nombre_cliente: str, carrito: Carrito,
                 cuenta_cliente: "BankAccount", cuenta_empresa: "BankAccount"):
        if not isinstance(nombre_cliente, str) or not nombre_cliente.strip():
            raise ValueError("El nombre del cliente debe ser un string no vacío.")
        if not isinstance(carrito, Carrito):
            raise TypeError("El parámetro 'carrito' debe ser una instancia de Carrito.")

        self.nombre_cliente = nombre_cliente.strip()
        self.carrito = carrito
        self.estado = "En proceso"

        # Nuevos atributos
        from_numbers = (int, float)  # por si usas tipos numéricos en account_number
        if not isinstance(cuenta_cliente, BankAccount):
            raise TypeError("cuenta_cliente debe ser una BankAccount.")
        if not isinstance(cuenta_empresa, BankAccount):
            raise TypeError("cuenta_empresa debe ser una BankAccount.")

        self.cuenta_cliente = cuenta_cliente
        self.cuenta_empresa = cuenta_empresa

    def procesar_pago(self):
        """
        Cambia el estado a 'Pagada' si el total del carrito es mayor que 0
        y se puede realizar la transferencia de la cuenta del cliente a la de la empresa.
        """
        total = self.carrito.calcular_total()

        if total <= 0:
            print("No se puede procesar el pago: el carrito está vacío.")
            return

        # Comprobamos saldo suficiente
        if self.cuenta_cliente.balance < total:
            print("No se puede procesar el pago: saldo insuficiente en la cuenta del cliente.")
            return

        # Hacemos la transferencia usando BankAccount.transfer
        self.cuenta_cliente.transfer(self.cuenta_empresa, total)
        self.estado = "Pagada"
        print(f"Pago procesado correctamente. Total: {total:.2f}€")
        print(f"Saldo cliente: {self.cuenta_cliente.balance:.2f}€")
        print(f"Saldo empresa: {self.cuenta_empresa.balance:.2f}€")

    def enviar_pedido(self):
        """Cambia el estado a 'Enviada' solo si el estado actual es 'Pagada'."""
        if self.estado == "Pagada":
            self.estado = "Enviada"
            print("El pedido ha sido enviado correctamente.")
        else:
            print("No se puede enviar: la orden aún no está pagada.")

    def __str__(self):
        """Devuelve información del cliente, estado y contenido del carrito."""
        return (
            f"Cliente: {self.nombre_cliente}\n"
            f"Estado: {self.estado}\n"
            f"{self.carrito}"
        )
```
